```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="muted")
```

```python
RESULTS_FILE = "../results/20260625-1903_benchmark.csv"

df = pd.read_csv(RESULTS_FILE)
df.head()
```

## Correctness vs Faithfulness by Run

```python
melted = df.melt(
    id_vars=["run"],
    value_vars=["correctness_pct", "faithfulness_pct"],
    var_name="metric",
    value_name="score",
)
melted["metric"] = melted["metric"].str.replace("_pct", "").str.replace("_", " ").str.capitalize()

fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=melted, x="run", y="score", hue="metric", ax=ax)
ax.set_xlabel("Run")
ax.set_ylabel("Score (%)")
ax.set_title("Benchmark: Correctness and Faithfulness by Run")
ax.set_ylim(0, 100)
ax.tick_params(axis="x", rotation=20)
plt.tight_layout()
plt.show()
```

## Correctness vs Faithfulness Scatter

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(
    data=df,
    x="correctness_pct",
    y="faithfulness_pct",
    hue="agent",
    style="agent",
    s=120,
    ax=ax,
)
for _, row in df.iterrows():
    ax.annotate(
        row["run"],
        (row["correctness_pct"], row["faithfulness_pct"]),
        textcoords="offset points",
        xytext=(6, 4),
        fontsize=8,
    )
ax.set_xlabel("Correctness (%)")
ax.set_ylabel("Faithfulness (%)")
ax.set_title("Correctness vs Faithfulness per Run")
ax.set_xlim(-5, 105)
ax.set_ylim(-5, 105)
plt.tight_layout()
plt.show()
```
